### Downloading and preparing dataset

We're using `google_images_download` module to download images directly from Google Graphics.

It will download images specified in queries into `data_dir/keyword` directories. If you already have your data you can skip this part (ctrl+/ to comment block of code).

````python
data_dir = base_dir + 'data'

# response = google_images_download.googleimagesdownload()
# queries = [
#     {
#         'keywords': 'hotdog',
#         'limit': 400,
#         'format': 'png',
#         'output_directory': data_dir,
#         'chromedriver': '/usr/bin/chromedriver'
#     },
#     {
#         'keywords': 'dachshund',
#         'limit': 400,
#         'format': 'png',
#         'output_directory': data_dir,
#         'chromedriver': '/usr/bin/chromedriver'
#     }]
# for args in queries:
#   img_path = response.download(args)
#   pprint(img_path)
````

We can verify if downloaded images are correct using `verify_images` from `fastai.vision`.

````python
doc(verify_images)
````

````python
classes = ['hotdog', 'dachshund']

for c in classes:
  print(c)
  folder = Path(data_dir)/c
  verify_images(folder, delete=True, max_size=500)
````

Let's create an ImageDataBunch - it prepares the data for the neural network learner.

````python
doc(ImageDataBunch)
````

````python
data = ImageDataBunch.from_folder(path=data_dir,             # path to the data (it will expect there a folder per class)
                                  train='.',                 # path to 'train' directory (we don't use this one)
                                  valid_pct=0.2,             # proportion of data that will be used for validation
                                  classes=classes,           # list of class names (data folder names)
                                  seed=42,                   # random seed
                                  size=224,                  # target size of every image
                                  num_workers=4              # number of CPU threads to be used
                                 ).normalize(imagenet_stats) # normalize using imagenet_stats
````

````python
data.show_batch(rows=3, figsize=(7,8))
````

### Creating and training CNN

We can create a Convolutional Neural Network model based on existing Resnet34 architecture.

Feel the might of Transfer Learning:

````python
doc(cnn_learner)
````

````python
learn = cnn_learner(data, models.resnet34, metrics=[accuracy, error_rate])
````

We can finally train our model.

Use `fit` method specifing number of epochs to run.

One epoch is one pass of the neural network through the whole training data.

````python
learn.fit(epochs=15)
````

````python
learn.recorder.plot_losses()
````

### Early stopping and model saving

The result is quite nice but we spend a lot of time training for nothing.

In the first phase the loss went down very rapidly, but then it hit the plateau. At the end it could even rise up.

We can use `EarlyStoppingCallback` that will stop the training when it doesn't improve (by `delta`) for a number of epochs (`patience`). \
`SaveModelCallback` can also help us saving a model everytime it improves.

````python
from fastai.callbacks import EarlyStoppingCallback, SaveModelCallback
doc(EarlyStoppingCallback)
doc(SaveModelCallback)
````

````python
learn = cnn_learner(data, models.resnet34, metrics=[accuracy, error_rate])
````

````python
early_stopping_cb = EarlyStoppingCallback(learn, monitor='accuracy', min_delta=0.001, patience=3)
save_model_cb = SaveModelCallback(learn, every='improvement', monitor='accuracy', name='bestmodel')

learn.fit(epochs=30, callbacks=[early_stopping_cb, save_model_cb])
````

````python
learn.recorder.plot_losses()
````

````python
learn.load('bestmodel')
print("Best model's accuracy on validation set:", learn.validate(data.valid_dl, metrics=[accuracy])[1].item())
````

### Unfreeze

Right now we are training only part of the model. Only last layers.

<img src="https://miro.medium.com/max/2000/1*f2_PnaPgA9iC5bpQaTroRw.png" alt="Transfer Learning" width=500/>

We can unfreeze previous layers and train the whole network. It doesn't always help so be carefull not to overwrite the previous model.

````python
learn.unfreeze()

save_model_cb = SaveModelCallback(learn, every='improvement', monitor='accuracy', name='bestmodel_unfreeze')

learn.fit(epochs=10, callbacks=[early_stopping_cb, save_model_cb])
````

````python
learn.recorder.plot_losses()
````

````python
learn.load('bestmodel'); freezed = learn.validate(data.valid_dl, metrics=[accuracy])[1].item()
learn.load('bestmodel_unfreeze'); unfreezed = learn.validate(data.valid_dl, metrics=[accuracy])[1].item()
print("Best freezed model accuracy:", freezed)
print("Best unfreezed model accuracy:", unfreezed)
````

### Interpretation

````python
interp.plot_confusion_matrix()
````

### Dataset transformations

Last time we made little changes to our dataset. But transformations can hugly impact our model.

Not only we can introduce more data to our model (which is crucial for Deep Learning) but also create harder examples for NN (like with poorer lighting) forcing it to overcome this obstacles by finding true patterns of presented objects. It really helps with overfitting.

It's called data augmentation.

<img src="https://www.researchgate.net/profile/Benoit_Gallix/publication/324457640/figure/fig1/AS:622298201595905@1525378861825/Graph-illustrating-the-impact-of-data-available-on-performance-of-traditional-machine.png" alt="Transfer Learning" width=500/>

````python
doc(get_transforms)
````

````python
tfms = get_transforms(do_flip=True,     # horizontal flip (mirror)
                      flip_vert=False,  # vertical flip
                      max_rotate=10,    # rotate image by degree
                      max_zoom=1.2,     # zoom image by factor
                      max_lighting=0.3, # change the lighting by factor
                      max_warp=0.2,     # warp the image by factor
                      p_affine=0.75,    # probability of transformation
                      p_lighting=0.75   # probability of lighting change
                     )
data = ImageDataBunch.from_folder(path=data_dir,
                                  train='.',
                                  valid_pct=0.2,
                                  seed=42,
                                  size=224,
                                  num_workers=4,
                                  classes=classes,
                                  ds_tfms=tfms # we added the transforms
                                 ).normalize(imagenet_stats)
data.show_batch(rows=3, figsize=(7,8))
````

### Improved training



For the previous model we used simple training wihout any additional features.

But there are plenty of tricks we can use to improve our training speed and quality. For example:
* Learning rate \
If our learning rate (lr) is to small it may take a long time to find a good result or we can even stay in a "local minimum". \
If lr is too large training can diverge and we will never find a minimum. \
We can introduce lr scanning and find optimal value of lr.
* [Fit "one cycle" method](https://sgugger.github.io/the-1cycle-policy.html) \
Recently proposed by Leslie Smith method not only speed up the convergance but also improve it's quality. \
At the beggining of the epoch we make the lr larger and larger and then, after reaching the specified maximum lr value, we decrement it.
* Learning rate slice \
After unfreezeing the model we give lower lr to the first layers of NN and higher to the last layers. \
It help us not chaning the weights drastically and diverging after unfreezing.

````python
learn = cnn_learner(data, models.resnet34, metrics=[accuracy, error_rate])
````

````python
learn.lr_find()
````

````python
learn.recorder.plot(suggestion=True)
min_grad_lr = learn.recorder.min_grad_lr
````

````python
early_stopping_cb = EarlyStoppingCallback(learn, monitor='accuracy', min_delta=0.001, patience=3)
save_model_cb = SaveModelCallback(learn, every='improvement', monitor='accuracy', name='bestmodel')

learn.fit_one_cycle(cyc_len=30, max_lr=min_grad_lr, callbacks=[early_stopping_cb, save_model_cb])
````

````python
learn.recorder.plot_losses()
````

````python
learn.load('bestmodel')
print("Best model's accuracy:", learn.validate(data.valid_dl, metrics=[accuracy])[1].item())
````

````python
learn.unfreeze()
learn.lr_find()
````

````python
learn.recorder.plot(suggestion=True)
min_grad_lr = learn.recorder.min_grad_lr
````

````python
save_model_cb = SaveModelCallback(learn, every='improvement', monitor='accuracy', name='bestmodel_unfreeze')

learn.fit_one_cycle(cyc_len=30, max_lr=slice(1e-5, 1e-4), callbacks=[early_stopping_cb, save_model_cb])
````

````python
learn.recorder.plot_losses()
````

````python
learn.load('bestmodel'); freezed = learn.validate(data.valid_dl, metrics=[accuracy])[1].item()
learn.load('bestmodel_unfreeze'); unfreezed = learn.validate(data.valid_dl, metrics=[accuracy])[1].item()
print("Best freezed model accuracy:", freezed)
print("Best unfreezed model accuracy:", unfreezed)
````

### Interpretation

````python
learn.load('bestmodel_unfreeze')
interp = ClassificationInterpretation.from_learner(learn)
````
